# One Hot Encoding - Variables With Many Categoies

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('test.csv', usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
data.head()
```

```python
# let's have a look at how many labels each variable has
data.nunique()
```

```python
# now do with using for loop
for col in data.columns:
    print(col , ':' ,len(data[col].unique()) , 'labels')
```

```python
# lets examine how many columns we will obtain after one hot encoding these variables
pd.get_dummies(data,drop_first=True).shape
```

```python
# lets find the top 10 most frequent categories for the variable x2
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# let make a list with the most frequent categories of the variables
top10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top10
```

```python
# now we make the 10 binary variables
for labels in top10:
    data[labels]= np.where (data['X2']==labels,1,0)
data[['X2']+ top10].head(40)    
```

```python
# get whole set of dummy variable for all the categorical variable
def one_hot_top_x(df,variable,top_labels):
    for labels in top_labels:
        df[variable+'_'+labels]= np.where(data[variable]==labels,1,0)
    
# read ur data again
data=pd.read_csv('test.csv', usecols=['X1','X2','X3','X4','X5','X6'])

one_hot_top_x(data,'X2',top10)
data.head()
```

```python
# find the 10 most frequent categoies for x1
top_10_x1 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10_x1
```

```python
one_hot_top_x(data,'X1',top_10_x1)
data.head()
```
